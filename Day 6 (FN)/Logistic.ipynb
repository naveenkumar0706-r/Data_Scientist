```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Example model creation
model = LogisticRegression(max_iter=1000)
```

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv(r"C:\Users\navee\Desktop\Data_Scientist\Day 5 (FN & AN)\insurance_data.csv")

# Display first 5 rows
print(df.head())

# Display dataset information
print("\nDataset Info:")
df.info()
```

```python
plt.scatter(df['age'], df['bought_insurance'], color='yellow')
plt.xlabel('Age')
plt.ylabel('Bought Insurance (0 = No, 1 = Yes)')
plt.title('Age vs Bought Insurance')
plt.grid(True)
plt.show()
```

```python
x = df[['age']]
y = df['bought_insurance']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

print("Predicted values:", y_pred.tolist)
print("Actual values:", y_test.tolist())
```

```python
import numpy as np

ages = np.linspace(20, 80, 100).reshape(-1, 1)
probs = model.predict_proba(ages)[:, 1]
plt.plot(ages.ravel(), probs, color='blue')
plt.xlabel('Age')
plt.ylabel('Probability of Buying Insurance')
plt.title('Logistic Regression Curve')
plt.grid(True)
plt.show()
```

```python
print("Will a 40-year-old buy insurance?")
print("Prediction (0=No, 1=Yes):", model.predict([[40]]))
print("Probability:", model.predict_proba([[40]])[0][1])
```
