```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
heights = [150,152,154,157,160,162,165,170,180,190,200,210,220,230,240]
print("Sample Height Data: ", heights)
```

```python
heights = np.array(heights).reshape(-1, 1)
print("Reshaped Heights for Clustering: \n", heights)
```

```python
KMeans = KMeans(n_clusters=3, random_state=42)
KMeans.fit(heights)
```

```python
labels = KMeans.labels_
centroids = KMeans.cluster_centers_
```

```python
print("Cluster Labels for each height: ", labels)
print("Centroids of the clusters: \n", centroids)
```

```python
# Step 4: Plot the clusters
plt.figure(figsize=(10, 4))
plt.scatter(heights, np.zeros_like(heights), c=labels, cmap='viridis', s=100)   #(150,0)
plt.scatter(centroids, [0,0,0] , c='red', s=200, marker='X', label='Centroids') #[0,0,0]
plt.title('K-Means Clustering of Student Heights')
plt.xlabel('Height (cm)')
plt.legend()
plt.grid(True)
plt.show()
```
